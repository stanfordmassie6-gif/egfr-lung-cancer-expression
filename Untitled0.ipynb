```python
import GEOparse
gse = GEOparse.get_GEO(geo="GSE32863", destdir="./")
```

```python
# Look at the structure of the dataset
print("Number of samples:", len(gse.gsms))
print("\nSample IDs:", list(gse.gsms.keys())[:5], "...")

# Peek at one sample's metadata
sample_id = list(gse.gsms.keys())[0]
sample = gse.gsms[sample_id]
print("\nFirst sample metadata:")
print(sample.metadata.keys())
print("\nSource/tissue type:")
print(sample.metadata.get('source_name_ch1'))
print("\nCharacteristics:")
print(sample.metadata.get('characteristics_ch1'))
```

```python
import pandas as pd

# Extract expression data and tissue type labels for all samples
expression_data = {}
labels = {}

for gsm_id, sample in gse.gsms.items():
    # Get tissue type label
    characteristics = sample.metadata.get('characteristics_ch1', [])
    tissue_type = None
    for char in characteristics:
        if char.startswith('type:'):
            tissue_type = char.replace('type:', '').strip()
            break
    labels[gsm_id] = tissue_type

    # Get expression values
    expression_data[gsm_id] = sample.table['VALUE']

# Build expression matrix (genes x samples)
expr_df = pd.DataFrame(expression_data)
expr_df.index = gse.gsms[list(gse.gsms.keys())[0]].table['ID_REF']

# Build metadata dataframe
meta_df = pd.DataFrame.from_dict(labels, orient='index', columns=['tissue_type'])

# Quick sanity check
print("Expression matrix shape:", expr_df.shape)
print("\nTissue type counts:")
print(meta_df['tissue_type'].value_counts())
```

```python
# Get the platform annotation to find EGFR probe IDs
platform_id = list(gse.gpls.keys())[0]
platform = gse.gpls[platform_id]

# Search for EGFR in the gene symbol column
annotation = platform.table
print("Annotation columns:", annotation.columns.tolist())

# Find EGFR probes
egfr_probes = annotation[annotation.apply(
    lambda row: row.astype(str).str.contains('EGFR', case=False).any(), axis=1
)]

print(f"\nFound {len(egfr_probes)} EGFR-related probes:")
print(egfr_probes[['ID', 'Symbol']].head(20) if 'Symbol' in annotation.columns
      else egfr_probes.head(20))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Extract the 4 direct EGFR probes
egfr_probe_ids = ['ILMN_1696521', 'ILMN_1728858', 'ILMN_1798975', 'ILMN_1755535']
egfr_expr = expr_df.loc[egfr_probe_ids]

# Average across probes (more robust than single probe)
egfr_mean = egfr_expr.mean(axis=0).to_frame(name='EGFR_expression')

# Merge with labels
egfr_mean = egfr_mean.join(meta_df)
egfr_mean['tissue_type'] = egfr_mean['tissue_type'].replace({
    'Adjacent non-tumor lung tissue': 'Normal',
    'Lung adenocarcinoma': 'Tumor'
})

# Statistical test
normal_vals = egfr_mean[egfr_mean['tissue_type'] == 'Normal']['EGFR_expression']
tumor_vals = egfr_mean[egfr_mean['tissue_type'] == 'Tumor']['EGFR_expression']

t_stat, p_val = stats.ttest_ind(tumor_vals, normal_vals)
print(f"Normal mean: {normal_vals.mean():.4f}")
print(f"Tumor mean:  {tumor_vals.mean():.4f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value:     {p_val:.6f}")

# Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Boxplot
sns.boxplot(data=egfr_mean, x='tissue_type', y='EGFR_expression',
            palette={'Normal': '#4C9BE8', 'Tumor': '#E8694C'}, ax=axes[0])
axes[0].set_title('EGFR Expression: Tumor vs Normal\n(GSE32863, Lung Adenocarcinoma)', fontsize=13)
axes[0].set_xlabel('')
axes[0].set_ylabel('Mean Expression (log2 intensity)')
axes[0].annotate(f'p = {p_val:.2e}', xy=(0.5, 0.95), xycoords='axes fraction',
                 ha='center', fontsize=11, color='black')

# Violin plot
sns.violinplot(data=egfr_mean, x='tissue_type', y='EGFR_expression',
               palette={'Normal': '#4C9BE8', 'Tumor': '#E8694C'}, ax=axes[1])
axes[1].set_title('EGFR Expression Distribution\n(Tumor vs Normal)', fontsize=13)
axes[1].set_xlabel('')
axes[1].set_ylabel('Mean Expression (log2 intensity)')

plt.tight_layout()
plt.savefig('EGFR_expression.png', dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved.")
```

## Results

EGFR expression was significantly elevated in lung adenocarcinoma tumor samples compared
to adjacent normal tissue (p = 1.32e-06, t-test, n=58 per group).

The means were relatively similar (7.36 vs 7.62 log2 intensity). Tumor samples showed
greater variance, consistent with known EGFR amplification and mutation
heterogeneity in NSCLC. High-expression outliers likely represent the EGFR-driven subtype
most responsive to EGFR-targeted therapies (e.g., osimertinib).

Data source: GSE32863 (NCBI GEO), Illumina microarray, 116 lung tissue samples.

## Discussion

EGFR overexpression in lung adenocarcinoma is well-established, and this analysis
confirms it in an independent dataset. The clinically challenging question is which tumors displaying EGFR overexpression are targetable for novel treatment, as EGFR expression does not constitute oncogenic activity on its own.
